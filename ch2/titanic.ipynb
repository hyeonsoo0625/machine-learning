```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
titanic_train = pd.read_csv('../ch1/train.csv')
titanic_test = pd.read_csv('../ch1/test.csv')
```

```python
titanic_train.head(10)
```

```python
titanic_train['Name'] = titanic_train['Name'].str.split(',').str[1].str.split('.').str[0]
titanic_train['Name'].head()
```

```python
titanic_train['Name'].value_counts()
```

```python
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
titanic_train['Name'] = le.fit_transform(titanic_train['Name'])
titanic_train['Name'].head(10)
```

```python
titanic_train.head()
```

```python
titanic_train['Sex'] = le.fit_transform(titanic_train['Sex'])
titanic_train.head()
```

```python
titanic_train.info()
```

```python
titanic_train['Age'].fillna(titanic_train['Age'].mean(), inplace=True)
titanic_train['Embarked'].fillna('N', inplace=True)
titanic_train['Cabin'].fillna('N', inplace=True)

titanic_train['Cabin'] = titanic_train['Cabin'].str[:1]
titanic_train.head()
```

```python
titanic_train['Cabin'] = le.fit_transform(titanic_train['Cabin'])
titanic_train['Cabin'].head()
```

```python
def changeAge(age):
    if age < 10: return 0
    elif age < 20: return 1
    elif age < 30: return 2
    elif age < 40: return 3
    elif age < 50: return 4
    elif age < 60: return 5
    elif age < 70: return 6
    else: return 7

titanic_train['Age'] = titanic_train['Age'].apply(lambda x : changeAge(x))
titanic_train.head()
```

```python
titanic_train['Embarked'] = le.fit_transform(titanic_train['Embarked'])
titanic_train.head(10)
```

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import cross_val_score
from sklearn.metrics import accuracy_score

X_train, X_test, y_train, y_test = train_test_split(titanic_train.drop('Survived', axis=1), titanic_train['Survived'], test_size=0.2)

dt_clf = DecisionTreeClassifier()
rf_clf = RandomForestClassifier()
lr_clf = LogisticRegression(solver='liblinear')

scores = cross_val_score(dt_clf, X_train, y_train, cv=5)

for clf in [dt_clf, rf_clf, lr_clf]:
    clf.fit(X_train, y_train)
    pred = clf.predict(X_test)
    accuracy = accuracy_score(y_test, pred)
    print(clf.__class__.__name__, accuracy)
```

```python
from sklearn.model_selection import GridSearchCV

parameters = {'max_depth':[2, 3, 5, 10],
              'min_samples_split':[2, 3, 5], 'min_samples_leaf':[1, 5, 8]}

grid_dclf = GridSearchCV(dt_clf, param_grid=parameters, scoring='accuracy', cv=5)
grid_dclf.fit(X_train, y_train)

best_dclf = grid_dclf.best_estimator_

dpredictions = best_dclf.predict(X_test)
accuracy = accuracy_score(y_test, dpredictions)
print(accuracy)
```
